# HuggingFace : des modèles entraînés

[HuggingFace](https://huggingface.co/) est un site recueillant les modèles IA entraînés, les plus connus (GPT, Bart, Mistral...) comme des plus modestes.
Ce site met aussi à disposition des dataset (comme le site kaggle).

Un modèle comprend plusieurs éléments. Par exemple, pour le texte, il contiendra un vocabulaire, l'outil de "tokenisation", un réseau entraîné, et un outil de restitution du résultat.

 
Les modèles peuvent être téléchargés en local (attention à la taille de certains), ou utilisés en ligne.

---
## Datasets
Hugging face propose un grand nombre de datasets dédié à l'apprentissage.
- Ces datasets, de taille variable, portent sur différents médias (texte, images, ....) et différentes  catégories (emails, CV, achats en ligne, santé, ...).

Il existe plusieurs façon d'accéder aux données : 
- les télécharger manuellement sur votre machine ou votre espace de travail
- les télécharger à partir d'un code python.

Il faut toutefois posséder un compte (gratuit, ou payant si vous souhaitez développer sur un espace hugging face avec une puissance de calcul).


### Créer un compte
Connectez vous sur [Hugging Face](https://huggingface.co/) et créez un compte si vous n'en avez pas déjà un.
 

### Charger un dataset
Prenons un dataset "léger" (110ko, 1500 lignes) et bien équilibré (50% de textes classés spam, 50% classés non spam), il est important d'avoir un bon équilibre pour l'apprentissage : "spam_notspam_dataset" de "Alauddin-Ali" (adresse [huggingface.co/datasets/Alauddin-Ali/spam_notspam_dataset](https://huggingface.co/datasets/Alauddin-Ali/spam_notspam_dataset))

Cherchons à y accéder via la librairie pandas.

```python
#il faut installer pandas avec pip install pandas
# et aussi installer datasets avec pip install datasets
# et pyarrow avec pip install pyarrow
# et fastparquet avec pip install fastparquet
```

```python
import pandas as pd
```

```python
from datasets import load_dataset
```

```python
dataset = load_dataset("Alauddin-Ali/spam_notspam_dataset")
```

```python
df = pd.DataFrame(dataset['train'])
```
